```python
## UPLOAD DATA
data_file = "/Users/rocfarriolduran/Desktop/BSC/B_SARS-CoV-2/A_SARS2_DEVELOPMENT/20211203_spike_omicron/I_final_candidates/20211210_omicron_final_dataframe.csv"
data = pd.read_csv(data_file, sep = ";")
data

## SORT ASCENDINGLY BY START
data_sorted = data.sort_values(by ='Start', ascending = 1)
data_sorted
```

### CONTIG EXTRACTOR AND SCORER

```python
    ### CONTIG, SCORE AND POSITION EXTRACTOR
    inici=0
    iniAnt=0
    fiAnt=0
    iniAct=0
    fiAct=0
    contig=""
    contig_vector=[]
    contig_vectors=[]
    Score_ABCpred=0.0
    Score_Bebipred_2_0=0.0
    Score_Discotope_2_0=-10.0
    #data = pd.read_csv(data_file, sep = ",")
    #Rank,Sequence,Start,End,Positions,Score,Tool,Length,extracellular1,Glycosilation,accessibility
    contig_vector=["Sequence","Start","End","Score_ABCpred","Score_Bebipred_2_0","Score_Discotope_2_0"]
    contig_vectors.append(contig_vector)
    for index, row in data_sorted.iterrows():
        iniAct=row[2]
        fiAct=row[3]
        if iniAct>=iniAnt and iniAct<=fiAnt and fiAct> fiAnt:
            contig=contig + row[1][fiAnt-fiAct:]
        elif iniAct>iniAnt and fiAct>fiAnt:
            if contig != "":
                contig_vector=[]
                contig_vector.append(contig)
                contig_vector.append(inici)
                contig_vector.append(fiAnt)
                contig_vector.append(Score_ABCpred)
                contig_vector.append(Score_Bebipred_2_0)
                contig_vector.append(Score_Discotope_2_0)
                contig_vectors.append(contig_vector)
                Score_ABCpred=0.0
                Score_Bebipred_2_0=0.0
                Score_Discotope_2_0=-10.0
            contig= row[1][:]
            inici=iniAct
        if row[6] == "ABCpred" and row[5]>Score_ABCpred:
            Score_ABCpred=row[5]
        if row[6] == "Bebipred 2.0" and row[5]>Score_Bebipred_2_0:
            Score_Bebipred_2_0=row[5]
        if row[6] == "Discotope 2.0" and row[5]>Score_Discotope_2_0:
            Score_Discotope_2_0=row[5]
        iniAnt=iniAct
        if fiAct > fiAnt:
            fiAnt=fiAct
    contig_vector=[]
    contig_vector.append(contig)
    contig_vector.append(inici)
    contig_vector.append(fiAnt)
    contig_vector.append(Score_ABCpred)
    contig_vector.append(Score_Bebipred_2_0)
    contig_vector.append(Score_Discotope_2_0)
    contig_vectors.append(contig_vector)
    for vector in contig_vectors:
        print(vector)
```
